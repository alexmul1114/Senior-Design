```python
import os
import json
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
import glob
```

```python
yololabel_path_train_HRSID = os.path.join(r"/home", "tjriz", "Documents", "Senior-Design", "datasets", "HRSID", "yolo_4", "train", "labels")
yoloimages_path_train_HRSID = os.path.join(r"/home", "tjriz", "Documents", "Senior-Design", "datasets", "HRSID", "yolo_4", "train", "images")
```

```python
labelfiles_train_HRSID = glob.glob(os.path.join(yololabel_path_train_HRSID, '*.txt'))
imagefiles_train_HRSID = glob.glob(os.path.join(yoloimages_path_train_HRSID, '*.png'))
print(len(labelfiles_train_HRSID))
print(len(imagefiles_train_HRSID))
```

```python
annotations = []
annotations_fileidx = []

for i in range(len(labelfiles_train_HRSID)):
    with open(labelfiles_train_HRSID[i], 'r') as file:
        for line in file:
            values = line.strip().split()
            class_id = int(values[0])
            x_center, y_center, width, height = map(float, values[1:])
            annotations.append([class_id, x_center, y_center, width, height])
            annotations_fileidx.append(i)
```

```python
count_occurences = 0
annotation_idx = []

for index, annotation in enumerate(annotations):
    norm_width, norm_height = annotation[3], annotation[4]
    if norm_width in (0, 1) or norm_height in (0, 1):
        count_occurences += 1
        annotation_idx.append(index)
        print(annotation)

print("\nOccurences of 0 or 1 normalized width/height values:", count_occurences)
```

```python
img_dim = 224

# Convert list to a NumPy array and extract width and height columns, scaling them to pixels:
ann_array = np.array(annotations)
dims_in_pixels = ann_array[:, 3:5] * img_dim

# For each bounding box, determine the larger pixel dimension:
largest_dims = dims_in_pixels.max(axis=1)

# Calculate statistics:
min_dim = largest_dims.min()
max_dim = largest_dims.max()
avg_dim = largest_dims.mean()

print("Minimum pixel dimension:", min_dim)
print("Maximum pixel dimension:", max_dim)
print("Average pixel dimension:", avg_dim)
```

```python
dataset_path_original = os.path.join(r"/home", "tjriz", "Documents", "Senior-Design", "datasets", "HRSID")
# annotation_path_original = os.path.join(dataset_path_original, "annotations", "train_test2017.json")
# image_path_original = os.path.join(dataset_path_original, "images")

def plot_original_json_bboxes(dataset_path, image_file_name):
    
    # Path to your original JSON annotation file
    annotations_file = os.path.join(dataset_path, 'annotations', 'train_test2017.json')

    # Load the JSON data
    with open(annotations_file, 'r') as f:
        data = json.load(f)

    # Lists of image metadata and their annotations
    images_info = data['images']
    annotations = data['annotations']

    # Find the image ID for the given file name
    image_id = None
    for img_info in images_info:
        if img_info['file_name'] == image_file_name:
            image_id = img_info['id']
            break

    if image_id is None:
        print(f"Image '{image_file_name}' not found in the JSON file.")
        return

    # Collect bounding boxes for that image
    bboxes = []
    for ann in annotations:
        if ann['image_id'] == image_id:
            # Each ann['bbox'] = [x, y, width, height]
            bboxes.append(ann['bbox'])

    # Open the image
    image_path = os.path.join(dataset_path, 'images', image_file_name)
    if not os.path.exists(image_path):
        print(f"Image file not found at path: {image_path}")
        return

    with Image.open(image_path) as img:
        # Convert to a Matplotlib figure
        fig, ax = plt.subplots(1, 1, figsize=(8, 8))
        ax.imshow(img)

        # Plot each bounding box
        for (x, y, w, h) in bboxes:
            rect = patches.Rectangle(
                (x, y), w, h,
                linewidth=2,
                edgecolor='red',
                facecolor='none'
            )
            ax.add_patch(rect)

        plt.title(f"{image_file_name}: {len(bboxes)} bounding box(es)")
        plt.axis('off')
        plt.show()

image_file_name = "P0131_13800_14600_9600_10400.png"

plot_original_json_bboxes(dataset_path_original, image_file_name)
```
